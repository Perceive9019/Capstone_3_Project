#### i. Data understanding

The columns in the dataset are:

- **SETTLEMET_DATE**: date in format dd/mm/yyyy
- **SETTLEMENT_PERIOD**: The half hourly period for the historic outtunr occurred. Settlement period 1 runs from 00:00-00:30.
- **ND (National Demand)**. National Demand is the sum of metered generation, but excludes generation required to meet station load, pump storage pumping and interconnector exports. National Demand is calculated as a sum of generation based on National Grid ESO operational generation metering. Measured in MW.
- **TSD (Transmission System Demand)**. Transmission System Demand is equal to the ND plus the additional generation required to meet station load, pump storage pumping and interconnector exports. Measured in MW.
- **ENGLAND_WALES_DEMAND**. England and Wales Demand, as ND above but on an England and Wales basis. Measured in MW.
- **EMBEDDED_WIND_GENERATION**. This is an estimate of the GB wind generation from wind farms which do not have Transmission System metering installed. These wind farms are embedded in the distribution network and invisible to National Grid ESO. Their effect is to suppress the electricity demand during periods of high wind. The true output of these generators is not known so an estimate is provided based on National Grid ESO’s best model. Measured in MW.
- **EMBEDDED_WIND_CAPACITY**. This is National Grid ESO’s best view of the installed embedded wind capacity in GB. This is based on publicly available information compiled from a variety of sources and is not the definitive view. It is consistent with the generation estimate provided above. Measured in MW
- **EMBEDDED_SOLAR_GENERATION**. This is an estimate of the GB solar generation from PV panels. These are embedded in the distribution network and invisible to National Grid ESO. Their effect is to suppress the electricity demand during periods of high radiation. The true output of these generators is not known so an estimate is provided based on National Grid ESO’s best model. Measured in MW.
- **EMBEDDED_SOLAR_CAPACITY**. As embedded wind capacity above, but for solar generation. Measured in MW.
- **NON_BM_STOR (Non-Balancing Mechanism Short-Term Operating Reserve)**. For units that are not included in the ND generator definition. This can be in the form of generation or demand reduction. Measured in MW.
- **PUMP_STORAGE_PUMPING**. The demand due to pumping at hydro pump storage units; the -ve signifies pumping load.
- **IFA_FLOW (IFA Interconnector Flow)**. The flow on on the respective interconnector. -ve signifies export power out from GB; +ve signifies import power into GB. Measured in MW.
- **IFA2_FLOW (IFA Interconnector Flow)**. The flow on the respective interconnector. -ve signifies export power out from GB; +ve signifies import power into GB. Measured in MW.
- **MOYLE_FLOW (Moyle Interconnector FLow)**. The flow on the respective interconnector. -ve signifies export power out from GB; +ve signifies import power into GB. Measured in MW.
- **EAST_WEST_FLOW (East West Innterconnector FLow)**. The flow on the respective interconnector. -ve signifies export power out from GB; +ve signifies import power into GB. Measured in MW.
- **NEMO_FLOW (Nemo Interconnector FLow)**. The flow on the respective interconnector. -ve signifies export power out from GB; +ve signifies import power into GB. Measured in MW.
- **is_holiday**. This data is in binary form representing two options presumably whether or not it is a holiday. i.e Bank Holidays, New Years etc.

### Bank Holidays
Bank holidays are important for accurate forecasting. For example, bank holidays can have a significant impact on time series analysis data, especially is related to human activit such as electricity consumption. Accounting for bank holidays makes more accurate forecasts and avoids underestimation or overestimating demand.

#### i. Compare Bank holidays in England and Wales to see if they are the same

```python
# Create an instance of holidays.UK class and call it bank_holidays_wales
# pass the parameters 'subdiv' set to England and 'years' set to a range of years
# Pass parameter 'observed' and set to True so observed bank holidays are included
#.items() method returns a list of tuples containing bank holiday and specified subdivision

wales_holidays = holidays.UK(subdiv='Wales', years=range(2009, 2024), observed=True).items()
```

Since the bank holidays are not the same, we are going to combine the bank holidays observed in both England and Wales, aligning bank holidays between the two regions.

```python
import holidays

# Create dictionaries of holidays for England and Wales
england_holidays = holidays.UK(subdiv='England', years=range(2009, 2024), observed=True)
wales_holidays = holidays.UK(subdiv='Wales', years=range(2009, 2024), observed=True)

# Combine the two dictionaries
combined_holidays = dict(list(england_holidays.items()) + list(wales_holidays.items()))

# Now, combined_holidays contains holidays from both England and Wales
```

```python
for date, name in combined_holidays.items():
    print(f'{date}: {name}')
```

#### Separate bank holidays (combined holidays) into two lists 'holiday_names' and 'holiday-dates' for regular holidays, and 'holiday_names_observed' and  'holiday_dates_observed' for observed holidays. Overwrite the observed regular holidays with the same data.

```python
# Create empty list to store data
holiday_names = []
holiday_dates = []
holiday_names_observed = []
holiday_dates_observed = []
```

```python
# Iterate through combined_holidays dictionary
for date, name in sorted(combined_holidays.items()):
    holiday_dates.append(date)
    holiday_names.append(name)
    
    # Pop the previous value as observed bank holidays take place later
    if "observed" in name:
        holiday_dates_observed.pop()
        holiday_names_observed.pop()
        
    holiday_names_observed.append(name)
    holiday_dates_observed.append(np.datetime64(date))
    
# print the first 5 observed holiday dates
print(holiday_dates_observed[:5])
```

These dates aligned with the dates in the combined_holidays dictionary

#### Filter out rows in the UK_data where the "tsd" column has a value of 0.0 by identifying the rows with matching "settlement_date" values and removing them from the DataFrame while resetting the index.

```python
# Identify unique settlement dates where 'tsd' is equal to zero
null_days = UK_data.loc[UK_data["tsd"] == 0.0, "settlement_date"].unique().tolist()

# Initialize empty list to store indices
null_days_index = []

# Iterate through unique settelment date and collect row indices
for day in null_days:
    # Find row indices where 'settlement_date' matches current date
    indices_for_day = UK_data[UK_data["settlement_date"] == day].index.tolist()
    
    # Append the indices to the null_days_index list
    null_days_index.append(indices_for_day)

# Flatten the nest list of indices
null_days_index = [item for sublist in null_days_index for item in sublist]

# Drop rows with the identified indices
UK_data.drop(index=null_days_index, inplace=True)

# Reset the index of the UK_data to be continous
UK_data.reset_index(drop=True, inplace=True)
```

### Feature creation
The first step in the feature creation is to change the date format to include the hourly values. The settlement_period values refer to how many samples have been taken per day. Given that there are 48 samples per day, each sample represents 30 minutes of the day. Using this information together with the valuable knowledge of the [StackOverflow](https://stackoverflow.com/questions/61444409/convert-a-pandas-column-to-hours-and-minutes) contributors

```python
UK_data.head(10)
```

```python
# Create a new column "period_hour" by converting "settlement_period" values to hours
UK_data['period_hour'] = (UK_data['settlement_period']).apply(
    lambda x: str(datetime.timedelta(hours=(x - 1) * 0.5))
)

# Replace "1 day, 0:00:00" with "0:00:00" in the "period_hour" column
UK_data.loc[UK_data["period_hour"] == "1 day, 0:00:00", "period_hour"] = "0:00:00"

# Move the "period_hour" column to a different position in the DataFrame
# First, remove it from its original location
move_old_column = UK_data.pop('period_hour')

# Then, insert it at the 2nd position (index 2)
UK_data.insert(2, "period_hour", move_old_column)

UK_data.sample(n=10)
```

```python
UK_data.head()
```

```python
UK_data["settlement_date"] = pd.to_datetime(
    (UK_data["settlement_date"] + " " + UK_data["period_hour"])
)

UK_data.sample(5)
```

### 3. Modelling

#### 3a. Decomposition

- What do we notice from the plot?
    - We can see that there is a trend downwards in Electricity demand. But there are also seasonal fluctuations. The seasonal flucatuations come in a repeated pattern. 

- We are going to decompose the UK_data. This means breaking it into 3 components:
    - **Trend**: The overall direction the data is traveling (upwards or downwards)
    - **Seasonality**: Cyclical patterns in the data
    - **Noise**: The randowm variation in the data
We can treat these components differently, depending on the question and what's appropriate in the context. They can either be added together in an additive model, or multiplied together in a multiplicative model.

```python
# Upload jpg file into Jupyter notebook
from IPython.display import Image
Image(filename='Additive or multiplicative?.jpg')
```

The UK_data is **multipicative** because it has increasing and decreasing amplitude and frequency over time. The magnitude of the seasonal pattern we see in the data is dependent on the magnitude of the data whereas in an **additive** model, the magnitude of seasonality does not change over time (changes over time are consistently made by the same amount, i.e. linear trend. Seasonality has the same amplitude and frequency).

We will use the decompose() function to quantify the component parts described above in our data.
