# UK Electricity Consumption Forecast - Time Series Analysis

The general framework we will use to carryout the time series forecasting analysis is as follows:

### 1. **Understanding the Dataset**
- Load our dataset into a dataset.
- Visualize the data to understand its pattern (e.g., using Matplotlib or Seaborn in Python). - Look for seasonality, trends, and potential outliers.
- Check for missing data and decide how we want to handle them (e.g., interpolation, forward-fill, or dropping).

### 2. **Decomposition of the Series**
Using the package `statsmodels.tsa.seasonal` in Python, we will decompose our time series into:
- Trend component
- Seasonal component
- Residual component

This gives us insights into the nature of the patterns in our data.

### 3. **Stationarity**
Most time series forecasting methods assume that the underlying time series is stationary, which means statistical properties like mean, variance, etc., are constant over time.

- Test for stationarity using the Dickey-Fuller test (`adfuller` function from `statsmodels.tsa.stattools`) and Kwiatkowski-Phillips-Schmidt-Shin test (`KPSS`function from `statsmodels.tsa.stattools`).
- If the series is non-stationary, we can make it stationary by differencing or taking the log, etc.

### 4. **Selecting a Model**
Bewlo are the models we will use to forecast electricity demand in the UK:

- **ARIMA** (AutoRegressive Integrated Moving Average): A popular model for time series forecasting. We'll need to find optimal parameters (p,d,q) for the model. We can use tools like `auto_arima` in the `pmdarima` library to find the best parameters.

- **Prophet**: Developed by Facebook, it's useful for datasets with strong seasonal patterns and multiple seasonality.

### 5. **Training and Validation**
- Split the dataset into a training set and a validation (or test) set. The training set will be used to train the model, while the validation set will be used to test its performance.
- Fit the selected model to the training data.
- Forecast for the validation period and compare the forecasted values with the actual values.

### 6. **Model Evaluation**
Use metrics like Mean Absolute Percentage Error (MAPE), Root Mean Squared Error (RMSE), or Mean Absolute Error (MAE) to measure the accuracy of our forecasts.

### 7. **Forecasting**
Once satisfied with the model's performance on the validation set, we can use the entire dataset to train the model and then forecast future consumption or demand.

### 8. **Post-processing**
Some models might give forecasts that need post-processing. For example, if we log-transformed our data, we'll need to reverse the transformation on our forecasts.

## 2. Decomposition of the Series
Given that we suspect seasonality in the data, decomposing the series into its trend, seasonal, and residual components can help validate this suspicion and provide deeper insights.

The data is **multipicative** because it has increasing and decreasing amplitude and frequency over time. The magnitude of the seasonal pattern we see in the data is dependent on the magnitude of the data whereas in an **additive** model, the magnitude of seasonality does not change over time (changes over time are consistently made by the same amount, i.e. linear trend. Seasonality has the same amplitude and frequency).

```python
# Upload jpg file into Jupyter notebook
from IPython.display import Image
Image(filename='Additive or multiplicative?.jpg')
```

We will compare models to determine which one provides a better fit.

**Additive Model**: The additive model is used when the seasonal variation is roughly constant throughout the series. In other words, the difference in amplitude between peaks and troughs remains approximately the same over time.

**Multiplicative Model**: The multiplicative model is appropriate when the seasonal variation increases or decreases as the time series data grows. In other words, as the data values increase, the difference between peaks and troughs becomes proportionally wider.

**Note**:
Since the data is collected every 30 minutes with 48 entries per day, and we're trying to capture daily seasonality, then the frequency should be set to 48 (since there are 48 data points in a single day).

However, if you're trying to capture another form of seasonality, such as weekly or yearly, you would adjust accordingly:

For weekly seasonality: 48 x 7 = 336

For yearly seasonality (assuming no leap year): 48 x 365 = 17,520

Given our data's granularity, it's common to first look at daily patterns, especially for data that is recorded at 30-minute intervals, as intraday variations might be of primary interest.

Thus, to capture daily seasonality, set period=48 when decomposing.

### 2.1 Decompose the series

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Decompose the tsd series
decomposition = seasonal_decompose(data['tsd'], model='additive', period=48)

# Plot the decomposed components
plt.figure(figsize=(15,10))

plt.subplot(4,1,1) # 4 rows, 1 column plot 1
plt.plot(data['settlement_date'], data['tsd'], label='Original')
plt.legend(loc='upper left')
plt.title('Original Electricity Demand (tsd) Series')

plt.subplot(4,1,2)
plt.plot(data['settlement_date'], decomposition.trend, label='Trend', color='blue' )
plt.legend(loc='upper left')
plt.title('Electricity Demand (tsd) Trend')

plt.subplot(4,1,3)
plt.plot(data['settlement_date'], decomposition.seasonal, label='Seasonal', color='green')
plt.legend(loc='upper left')
plt.title('Electricity Demand (tsd) Seasonal')

plt.subplot(4,1,4)
plt.plot(data['settlement_date'], decomposition.resid, label='Residual', color='red')
plt.legend(loc='upper left')
plt.title('Electricity Demand (tsd) Residual')

plt.tight_layout()
plt.show()
```

### Why is the Seasonal Plot a solid green reactangle?

The seasonal plot displays as a solid green rectangle suggesting that the amplitude of the seasonal component is so large in comparison to the other plotted values that it essentially dwarfs them, making the plot look like a solid block.

There are a few possible reasons and solutions for this:

- **Scale Differences**: The seasonal variations might be very large compared to the trend and residual components. This could cause the seasonal plot to appear as a solid block when plotted on the same scale as the other components. In such cases, we can plot the seasonal component on a different scale.

- **High Frequency**: If we're capturing daily seasonality with period=48 and have several years of data, the result will be a very high-frequency seasonal pattern. When plotted over several years, this can appear as a solid block because the fluctuations happen so frequently that they're hard to distinguish at the scale of the entire dataset.

- **Decomposition Model**: As previously discussed, we might be using the wrong decomposition model (additive vs. multiplicative). If the seasonal pattern is better described by a multiplicative model but we're using an additive model, the seasonal decomposition might not be accurate.

```python
# Decompose the tsd series
decomposition = seasonal_decompose(data['tsd'], model='multiplicative', period=48)

# Plot the decomposed components
plt.figure(figsize=(15,10))

plt.subplot(4,1,1) # 4 rows, 1 column plot 1
plt.plot(data['settlement_date'], data['tsd'], label='Original')
plt.legend(loc='upper left')
plt.title('Original Electricity Demand (tsd) Series')

plt.subplot(4,1,2)
plt.plot(data['settlement_date'], decomposition.trend, label='Trend', color='blue' )
plt.legend(loc='upper left')
plt.title('Electricity Demand (tsd) Trend')

plt.subplot(4,1,3)
plt.plot(data['settlement_date'], decomposition.seasonal, label='Seasonal', color='green')
plt.legend(loc='upper left')
plt.title('Electricity Demand (tsd) Seasonal')

plt.subplot(4,1,4)
plt.plot(data['settlement_date'], decomposition.resid, label='Residual', color='red')
plt.legend(loc='upper left')
plt.title('Electricity Demand (tsd) Residual')

plt.tight_layout()
plt.show()
```

**Note**:
The multiplicative model also shows the seasonal plot as a solid green block.

## 3. Testing for Stationarity
Before you can apply most time series forecasting methods, you need to ensure that your time series is stationary. Stationarity means that the statistical properties of the series, like mean, variance, and autocorrelation, are constant over time.

### 3.1 ADF Test: 
The Augmented Dickey-Fuller test is a formal statistical test for stationarity. The null hypothesis of the test is that the time series can be represented by a unit root, which is not stationary (has some time-dependent structure). The alternative hypothesis is that the time series is stationary.

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(data['tsd'])

# Pring test-statistic
print('ADF Statistic: %f' % result[0])

#Print p-value
print('p-value: %f' % result[1])
```

The Augmented Dickey-Fuller (ADF) test results indicate the following:

 - ADF Statistic: A value of -16.920976 is highly negative.
 - p-value: A value of 0.000000 (essentially 0) is far below the common significance level (like 0.05).
 
**Interpretation**:

Given these results, you would reject the null hypothesis of the ADF test, suggesting that the time series does not have a unit root and is stationary.

    - Remember:
    - Null Hypothesis (H0): The time series has a unit root (i.e., it's non-stationary).
    - Alternative Hypothesis (H1): The time series does not have a unit root (i.e., it's stationary).

This is a good indication, especially if we're looking to model oour time series with methods like ARIMA or SARIMA. A stationary series often leads to better model performance because the statistical properties (mean, variance, autocorrelation) of the series remain constant over time, which is an assumption in many time series forecasting methods.

Given that our data appears to be stationary, we can proceed to the next steps of modeling, which involve selecting an appropriate time series forecasting model and training it on your data.

### 3.2 Testing for stationarity with KPSS
Checking for stationarity using both the ADF and KPSS tests provides a more comprehensive view of the data's stationarity.

**KPSS Test** :
- Null Hypothesis (H0): The time series is stationary around a deterministic trend.
- Alternative Hypothesis (H1): The time series is not stationary (has a unit root).

     - If the p-value is less than a threshold (e.g., 0.05), then you reject the null hypothesis, suggesting the time series is non-stationary.
    - If the p-value is greater than the threshold, you fail to reject the null hypothesis and infer that the time series is stationary.


```python
from statsmodels.tsa.stattools import kpss

# Perform the KPSS test
kpss_result = kpss(data['tsd'])

# Print the test-statistic
print('KPSS Statistic: %f' % kpss_result[0])

# Print the p-value
print('p-value: %f' % kpss_result[1])

# Print Critical Values
print('Critical Values:', kpss_result[3])
```

**Note**:
For the KPSS test, the critical values determine the thresholds at which you can reject the null hypothesis. The KPSS test generally provides critical values for various significance levels (e.g., 1%, 5%, 10%).

If the KPSS statistic is greater than the critical value at a certain significance level, then you can reject the null hypothesis at that significance level.

The KPSS test results indicate the following:

- **KPSS Statistic**: A value of 41.546408 is significantly larger than any of the provided critical values.
- **p-value**: A value of 0.010000 (or 1%) is below the common significance levels (like 0.05 or 0.01).

**Interpretation**:

Given these results:

1. The KPSS statistic of 41.546408 is well above all the critical values, including the 1% level.
2. The p-value of 0.010000 suggests that you reject the null hypothesis at a 1% significance level.

This means that according to the KPSS test, we would reject the null hypothesis, suggesting that the time series **is not stationary around a deterministic trend**.

However, remember that the ADF test indicated that the series was stationary. This discrepancy between the ADF and KPSS tests suggests that the series might be **difference-stationary**. In other words, differencing the series once might make it stationary.

In practice, when you get conflicting results from ADF and KPSS:

- ADF indicates stationary, and KPSS indicates non-stationary: The series might be difference-stationary. Differencing can make the series stationary, and you might consider using techniques like ARIMA that involve integrated (I) components.

Given the results, the next logical step could be to difference the series once and then re-run both the ADF and KPSS tests on the differenced data to see if it becomes stationary. Alternatively, you could proceed with models that can handle and account for non-stationarity, like SARIMA or Prophet.

**Note**:
Difference-stationary means that the 'tsd' series becomes stationary after differencing it a certain number of times. The most common approach is to take the first difference, but sometimes a second or even third difference might be necessary.

### 3.3 Compute Difference-Stationary

```python
# Create a new column called tsd_diff
data['tsd_diff'] = data['tsd'].diff().dropna() # drop missing values
```

```python
adf_result_diff = adfuller(data['tsd_diff'].dropna())
print('ADF Statistic (Difference Data):', adf_result_diff[0])
print('p-value:', adf_result_diff[1])
```

```python
kpss_result_diff = kpss(data['tsd_diff'].dropna())
print('KPSS Statistic (Difference Data):', kpss_result_diff[0])
print('p-value:', kpss_result_diff[1])
print('Critical Values:', kpss_result_diff[3])
```

Let's interpret these results:

##### ADF Test (on Differenced Data):

- **ADF Statistic**: \(-66.06649322169788\) is highly negative.
- **p-value**: 0.0 is essentially 0.

**Interpretation**: The ADF test suggests that the differenced series is stationary, as the p-value is significantly less than the commonly used significance level (0.05).

##### KPSS Test (on Differenced Data):

- **KPSS Statistic**: \(0.0007534262677409369\) is way below the critical values.
- **p-value**: 0.1 is above the common significance levels.

**Interpretation**: Given the p-value(<0.005) and the KPSS statistic being below the critical values, we fail to reject the null hypothesis. This suggests that the differenced series is stationary around a deterministic trend.

In conclusion, both tests agree that after taking the first difference, our time series data is stationary. This is good news for most time series forecasting techniques, as many of them require or at least perform better when the data is stationary.

The next step involves choosing and fitting a forecasting model to the differenced (stationary) data, like ARIMA or SARIMA. We can also work on selecting the right order of differences, AR terms, and MA terms if you'd like to proceed with the ARIMA model.

## 4. Selecting a Model

Since our data has both a seasonal component and has been differenced to achieve stationarity, a Seasonal ARIMA (SARIMA) model would be a suitable choice. SARIMA allows for both non-seasonal (p, d, q) and seasonal (P, D, Q, S) components.

### 4.1 Model Order Selection
To determine the optimal Non-seasonal orders: (p, d, q) and Seasonal orders:(P, D, Q, S) parameters, we can utilize tools like the auto_arima function from the pmdarima library, which conducts a search over possible model orders and chooses the best.

**Note**:
- Non-seasonal orders                                  
     - p:autoregressive order
     - d:differencing order
     - q:moving averge order
     
- Seasonal orders                                  
     - P:seasonal autoregressive order
     - D:seasonal differencind order
     - Q:seasonal moving averge order   
     - S: number of time steps per cycle

**Note:**

We will use the pmdarima: This is a third-party library that provides the auto_arima function, among other utilities. auto_arima helps in determining the best ARIMA/SARIMA parameters (p, d, q, P, D, Q, S) by trying out a range of values and selecting the one with the lowest AIC (or other specified criteria). Essentially, it's a useful tool for hyperparameter tuning for ARIMA/SARIMA models.

```python
from pmdarima import auto_arima

# Determine the SARIMA order for the differenced data
sarima_order = auto_arima(data['tsd_diff'].dropna(), seasonal=True, m=48, trace=True,
                          error_action='ignore', suppress_warnings=True, stepwise=True, 
                          d=0)  # Set d=0 because data is already differenced

print(sarima_order.summary())
```

**Note**: auto_arima parameters

- **seasonal=True**: Indicates we want to consider the seasonal component.
- **m=48**: This specifies the seasonal frequency of the time series (30 min intervals/day, in this case).
- **trace=True**: This will show progress during the search.
- **stepwise=True**: Accelerates the search process by considering a more restricted set of possible parameter combinations (based on the best parameters found in the previous step).

Once you've run the above code, the summary will provide the order (p, d, q) and seasonal order (P, D, Q, S) that were found to be the best for your dataset.
